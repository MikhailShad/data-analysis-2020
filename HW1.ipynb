```python
import pandas as pd

raw_data = pd.read_stata("data_games.dta")
raw_data
```

```python
#Task 1.1
payments_by_giftcard = len(raw_data[raw_data.payment_method == 'giftcard'])
print("How many payments were done using a giftcard? {}".format(payments_by_giftcard))

all_payments = len(raw_data)

payments_by_fb_promotion = len(raw_data[raw_data.payment_method == 'fb_promotion'])
percent_of_fb_promotion = 100 * payments_by_fb_promotion / all_payments
print("What is the percent of payments, which were done using fb_promotion? {}".format(percent_of_fb_promotion))

percent_of_giftcards = 100 * payments_by_giftcard / all_payments
print("What is the percent of payments, which were done using giftcards? {}".format(percent_of_giftcards))
```

```python
#Task 1.2
import numpy as np
from scipy import stats

crystalls_bought_series = raw_data['crystalls_bought']

mode = stats.mode(crystalls_bought_series)
print("Mode = {}".format(mode))

median = np.median(crystalls_bought_series)
print("Median = {}".format(median))

mean = np.mean(crystalls_bought_series)
print("Mean = {}".format(mean))

d_range = np.max(crystalls_bought_series) - np.min(crystalls_bought_series)
print("Range = {}".format(d_range))

std = np.std(crystalls_bought_series)
print("Std = {}".format(std))

s_e_mean = std / np.sqrt(len(crystalls_bought_series))
print("S. E. mean = {}".format(s_e_mean))

interquartile_range = crystalls_bought_series.quantile(0.75) - crystalls_bought_series.quantile(0.25)
print("Interquartile Range = {}".format(interquartile_range))

quartile_deviation = interquartile_range / 2
print("Quartile Deviation = {}".format(quartile_deviation))

decile_ratio = crystalls_bought_series.quantile(0.9) / crystalls_bought_series.quantile(0.1)
print("Decile ratio = {}".format(decile_ratio))
```

```python
import matplotlib.pyplot as plt

plt.hist(crystalls_bought_series, 25, facecolor='blue', alpha=0.5)
plt.show()
```
