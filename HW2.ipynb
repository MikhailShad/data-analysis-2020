```python
import pandas as pd

#Task 1
raw_data = pd.read_stata("salary.dta")
raw_data
```

```python
#Task 1.1
raw_data.position = raw_data.position.astype('category').cat.codes
raw_data
```

```python
#Task 1.2
usdrub = 63.472 #Данные взяты с https://www.finam.ru/profile/kurs-rublya/usd-from-cb/
def calculate_taxes_in_usd(row):
    taxes = row.salary * 0.3 if row.foreigner == 'yes' else row.salary * 0.13
    return taxes / usdrub

raw_data.taxes = raw_data.apply(calculate_taxes_in_usd, axis=1)
raw_data
```

```python
#Task 1.3
raw_data.salary_rank. = raw_data.salary.rank(ascending=False, method='dense')
raw_data
```

```python
#Task 1.4
raw_data.sort_values('salary', inplace=True, ascending=False)
raw_data
```

```python
#Task 1.5
raw_data.to_stata("salary.dta")
```

```python
#Task 2
raw_data = pd.read_stata("data_games.dta")
raw_data
```

```python
#Task 2.1
raw_data.drop(raw_data[(raw_data.payment < 100) | (raw_data.payment > 14000)].index, inplace=True)
raw_data
```

```python
#Task 2.2
def group_payment(row):
    if row.payment > 2000:
        return 4
    elif row.payment > 1000:
        return 3
    elif row.payment > 500:
        return 2
    elif row.payment > 100:
        return 1

raw_data['group'] = raw_data.apply(group_payment, axis=1)
print('The mode of group is {}'.format(raw_data.group.mode()[0]))
raw_data
```

```python
#Task 3
raw_data = pd.read_stata('revenue.dta')
print(raw_data)

mean_revenue = pd.DataFrame({'mean_revenue': raw_data.groupby(['company','year'])["rev"].mean()}).reset_index()
mean_revenue
```

```python
#Task 4
raw_data = pd.read_stata("data_games.dta")
print(raw_data)

contingency_table  = pd.crosstab(raw_data.payment_method, raw_data.country) 
contingency_table
```

```python
#Task 4.1
print('The most frequent payment method used in Russia: {}'.format(contingency_table.RU.idxmax()))
```

```python
#Task 4.2
transposed_contingency_table = contingency_table.T
gift_card_payments = transposed_contingency_table.giftcard.sum()

print('The percentage of gift card from all the worldwide payments: {}'.format(100 * gift_card_payments / len(raw_data)))
```

```python
#Task 4.3
print('Players from {} made most payments using mobile payment type'.format(transposed_contingency_table.mobile.idxmax()))
```
